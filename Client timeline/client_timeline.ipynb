```
import pandas
from datetime import date, datetime
from workalendar.asia import HongKong
cal = HongKong()
```

```
def calculate_this_quarter_date(last_quarter_date):
  if last_quarter_date.month == 6:
    nu_working = cal.get_working_days_delta(date(2020, 6, 1), last_quarter_date)
    this_quarter_date = cal.add_working_days(date(2020, 9, 1), nu_working)
  if last_quarter_date.month > 6:
    nu_working = cal.get_working_days_delta(date(2020, 7, 1), last_quarter_date)
    this_quarter_date = cal.add_working_days(date(2020, 10, 1), nu_working)
  return this_quarter_date 
```

```
last_quarter_date = pandas.read_excel("/content/sample_data/Client timeline work.xlsx", sheet_name="Sheet2", header=None).applymap(lambda x: datetime.strptime(str(x)[:10], "%Y-%m-%d"))
last_quarter_date.columns = ["last_quarter_date"]
```

```
last_quarter_date
```

```
last_quarter_date["this_quarter_date"] = last_quarter_date["last_quarter_date"].apply(calculate_this_quarter_date)
last_quarter_date
```

```
last_quarter_date.to_csv("/content/sample_data/Client timeline work.csv", index=False, header=True)
```

```
last_quarter_date["is_workday"] = last_quarter_date["this_quarter_date"].apply(cal.is_working_day)
```
